```
#list
import numpy as np
my_array = np.array([1, 2, 3, 4, 5])
print("First element:", my_array[0])
print("Last element:", my_array[-1])
my_array[2] = 10
print("Modified list:", my_array)

#array
import numpy as np
array_a = np.array([1, 2, 3])
array_b = np.array([4, 5, 6])

multiplication = array_a * array_b
print("Multiplication:", multiplication)

subtraction = array_a - array_b
print("Subtraction:", subtraction)

# Identify their type using type()
import numpy as np
import pandas as pd
array_1d = np.array([1, 2, 3, 4, 5])
print("Type of array_1d:", type(array_1d))

array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("Type of array_2d:", type(array_2d))

series = pd.Series([1, 2, 3, 4, 5])
print("Type of series:", type(series))

dataframe = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
print("Type of dataframe:", type(dataframe))

#Perform mathematical operations on these datasets created - multiplication, division,poweroff

import numpy as np
array_a = np.array([1, 2, 3, 4, 5])
array_b = np.array([5, 4, 3, 2, 1])

print("Array A:", array_a)
print("Array B:", array_b)
#  multiplication
multiplication_result = array_a * array_b
print("Multiplication Result:", multiplication_result)
# division
division_result = array_a / array_b
print("Division Result:", division_result)

#Combine text with Numpy fuction to generate a textual output - ”Addition of Two:array1 +array 2
import numpy as np
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
result = array1 + array2
output = f"Addition of Two:\n{array1} + {array2} = {result}"
print(output)

#Implement np.sin(), log(), log2(), np.exp())

import numpy as np
x = np.array([0, np.pi/6, np.pi/4, np.pi/3, np.pi/2])
print("Array of angles in radians:", x)

# 1. Sine function
sine_values = np.sin(x)
print("Sine values:", sine_values)

# 2. Natural logarithm
y = np.array([1, np.e, 10, 100])
print("\nArray for logarithmic functions:", y)
natural_log_values = np.log(y)
print("Natural logarithm values:", natural_log_values)

# 3. Base-2 logarithm
log2_values = np.log2(y)
print("Base-2 logarithm values:", log2_values)

# 4. Exponential function
exp_values = np.exp(y)
print("Exponential values:", exp_values)
```

```
#Generate a Data set for Health Care using the following:people = [’kiran’, arun’, ’vijay’, ’varun’]age =[25, 30, 35, 40, 45]height =[145, 151, 165, 173]weight=[45, 55, 65, 75]

import numpy as np
import pandas as pd
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

data = pd.DataFrame({
    'Person': people,
    'Age': age,
    'Height': height,
    'Weight': weight
})

print(data)

#Using the generated dataset given above. Generate the following graphs and justify therelationships among the vectors

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(data['Height'], data['Weight'], color='blue', marker='o')
plt.title('Scatter Plot of Height vs Weight', fontsize=14)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)
plt.grid()
plt.show()


# Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(data['Person'], data['Weight'], color='pink')

plt.title('Bar Chart of Weights of Individuals', fontsize=14)
plt.xlabel('Person', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)
plt.grid(axis='y')
plt.show()


# Histogram
plt.figure(figsize=(8, 5))
plt.hist(data['Weight'], bins=4, color='violet', edgecolor='red')
plt.title('Histogram of Weights', fontsize=14)
plt.xlabel('Weight (kg)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y')
plt.show()
```
